```python
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest
```

```python
from scipy.stats import ttest_ind, levene, shapiro
import pandas as pd
df = pd.read_csv('data/service_requests.csv')
df = pd.read_csv('data/filtered_data.csv')
# Filter data for the two departments
eng_response = df[df['department'] == 'ENG']['response_time']
dbl_response = df[df['department'] == 'DBL']['response_time']

# Check normality
print("Shapiro-Wilk Test for Normality:")
print("ENG:", shapiro(eng_response))
print("DBL:", shapiro(dbl_response))

# Check variance equality
print("Levene's Test for Equal Variance:")
print(levene(eng_response, dbl_response))

# Perform two-sample t-test
t_stat, p_value = ttest_ind(eng_response, dbl_response, equal_var=True)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
